```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
```

```python
import ast
import re
def fetch_name(data):
    words=[]
    for obj in ast.literal_eval(data):
        words.append(obj['name'])

    filtered_words=[]
    for word in words:
        cleaned_string = re.sub(r'[^a-zA-Z0-9\s]', '', word).lower()
        csl = cleaned_string.split(" ")
        for w in csl:
            if w not in stop_words and len(w)>=2:
                filtered_words.append(w)

    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

    return lemmatized_words
    # return(l)
```

```python
def process_sentence(sentence):
    # print(sentence)
    words = word_tokenize(sentence)
    # filtered_words = [word for word in words if word.lower() not in stop_words]
    filtered_words=[]
    for word in words:
        cleaned_string = re.sub(r'[^a-zA-Z0-9\s]', '', word).lower()
        csl = cleaned_string.split(" ")
        for w in csl:
            if w not in stop_words and len(w)>=2:
                filtered_words.append(w)

    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

    return lemmatized_words
```

```python
movies['genres']=movies['genres'].apply(fetch_name)
movies['keywords']=movies['keywords'].apply(fetch_name)
```

```python
movies['overview']=movies['overview'].apply(process_sentence)
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
def word2vec_on_dataframe(arr):
    vec = np.zeros(300)
    count=0
    for word in arr:
        try:
            vec+=google_word2vec[word]
            count+=1
        except:
            continue
    if count>0:
        vec/=count
    return vec
```

```python
new_df["word2vec"] = new_df["tags"].apply(word2vec_on_dataframe)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
def recommend(movie):
    movie_row =new_df[new_df['title']==movie]

    # movie_id = movie_row["movie_id"]
    movie_id = movie_row.iloc[0]['movie_id']
    # print(movie_id)
    movie_vec = movie_row.iloc[0]["word2vec"]
    similarities=[]
    for index, row in new_df.iterrows():
        id,name,vec = row["movie_id"],row["title"],row["word2vec"]
        similarity = cosine_similarity([movie_vec],[vec])[0][0]
        similarities.append([similarity,id,name])
    similarities=sorted(similarities)[::-1]

    return similarities[:10]
```

```python
sim = recommend("The Incredible Hulk")
```

```python
for similarity, id, title in sim:
  print(title, id, similarity)
```
